```python
import os
import sys
root_dir = os.getcwd()
sys.path.append(root_dir + '/app')
```

```python
from app.driver import read_csv, extract_tweeted_urls, CustomDriver
from app.publication import Publication
```

```python
tweets = read_csv('example_data.csv')
urls = extract_tweeted_urls(tweets)

wsj_urls = urls.loc[10:15]
wsj_urls
```

```python
driver = CustomDriver(root_dir + '/chromedriver')

wsj_kwargs = {
    'images': [{
        'parent_div': 'div.wsj-article-caption',
        'caption_span': 'span.wsj-article-caption-content',
        'credit_span': 'span.wsj-article-credit'}],
    'date': 'time.timestamp'
}

wsj = Publication('WSJ', cookies={}, kwargs=wsj_kwargs)
wsj.articles = driver.scrape_articles(wsj, wsj_urls)
wsj.write('notebook_output.json')  # write output data
driver.quit()
```
